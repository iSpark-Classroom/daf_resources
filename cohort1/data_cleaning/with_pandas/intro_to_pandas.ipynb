### Pandas Data Structures
- Series: a one-dimensional labeled array holding data of any type such as integers, strings, Python objects etc. Think of it like a column in a table.
- DataFrame: a two-dimensional data structure that holds data like a two-dimension array or a table with rows and columns.

```python
import numpy as np
s = pd.Series([1, 3, 5, np.nan, 6, 8])
```

```python
s
```

In pandas, np.nan stands for Not a Number. It is a special floating-point value inherited from the NumPy library and is used as the standard marker for missing or undefined data.

Conceptually, `np.nan` is equivalent to `None` and to `pd.NA` as they are all markers for the absence of a value although they all have their unique properties and ways of working with it.

```python
df2 = pd.DataFrame(
    {
        "A": 1,
        "B": pd.Timestamp("20130102"),
        "C": pd.Series(1, list(range(4)), dtype="float32"),
        "D": np.array([3] * 4, dtype="int32"),
        "E": pd.Categorical(["test", "train", "test", "train"]),
        "F": "foo",
    }
)
df2
```

```python
sr = pd.Series(
    [1, 2, 3, 4, 5, 6],
    index=['Q', 'W', 'E', 'R', 'T', 'Y'],
    name="Number List"
)
sr
```

We can locate a row using the `loc` or `iloc` methods. The loc is used with labls and iloc is used with positions. Hence the names `loc` -> location and `iloc` -> integer location.

Here are some examples below:

```python
sr.loc["Q":"R"]
```

```python
sr.iloc[0:3]
```

```python
sr2 = [1, 2, 3]
sr2 = pd.Series(sr2)
```

```python
sr2.index = [11, 12, 13]
sr2
```

```python
sr2.loc[11]
```

```python
sr2.iloc[0]
```

### DataFrames Deep-dive

```python
df = pd.DataFrame({
    "Age": [21,19,22,20,23,23,21],
    "Sex": ["M","M","F","M","F","F","M"],
    "GPA": [3.45,2.98,3.72,2.87,3.90,4.00,1.90]
}, index=["James", "Mark", "Rebecca", "David", "Lucy", "Judy", "Johnny"])

df
```

```python
df.head()
```

```python
df.tail()
```

What do you notice?

Calling our DataFrame or Series with .head() will return the first rows of our dataset. Passing an int to this function will return the first nth rows, but will return the first 5 rows as default (if you don't pass a number). However, if you want to access the last rows, you can use the .tail() function.
